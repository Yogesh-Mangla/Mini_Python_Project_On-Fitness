NPS : NET PROMOTOR SCORE:

1-6 RATING : DEMOTER/DTRACTOR
7-8 RATING : NEUTRAL
9-10 RATING : PROMOTORS


NPS = NO OF PROMOTORS - NO OF DETRACTORS/ TOTAL USERS


COMPANY DATASET  HAVING DATA ON CUSTOMER SATISFACTION AND CALCULATE NPS USING NUMPY


```python
import numpy as np
```

```python
survey = np.loadtxt('/content/survey.txt')
survey[:5]
```

```python
#What nupy technique to be used to filter out detractors
```

```python

detractors = survey[survey <=6]
```

```python
detractors
```

```python
detractors.shape
```

Find promotor percentage [promotors > =9]

```python

promoters = survey[survey >=9]
#promoter_percent = np.mean(promoters)*100
promotor_percent = round(promoters.shape[0]/survey.shape[0]*100,2)
```

```python

#promoter_percent = np.mean(promoters)*100
detractor_percent = round(detractors.shape[0]/survey.shape[0]*100,2)
```

```python
#NPS = PERCENTAGE OF (PROMOTORS) - PERCENTAGE OF (DETRACTORS)
```

```python
NPS = promoter_percent - detractor_percent
```

```python
NPS
```

```python
NPS = round(100*(promoters.shape[0]-detractors.shape[0])/survey.shape[0],2)
```

```python
print("NPS = ",NPS,"%")
```

```python
#What is the longest streak (in days) where 'Inactive' days had calories burned less than the daily mean across the full dataset?
```

```python
max_streak = 0
current_streak = 0

for i in conditions:
  if i :
    current_streak+=1
    max_streak = max(max_streak,current_streak)
  else :
    current_streak = 0
```
